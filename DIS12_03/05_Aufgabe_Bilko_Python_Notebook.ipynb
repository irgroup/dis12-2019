```python
import os, re
import glob
import math
import pandas as pd



#search_term = input("Bitte geben Sie den Suchterm ein: ");



def document_analyser(list_of_searchTerms,tf_boolean,df_boolean,idf_boolean,tf_idf_boolean,frequency_boolean,binary_boolean):
    main_dataframe = {('', ''): {}}
    for i in list_of_searchTerms:
        
        
        
        pattern = re.compile(''+i+'', re.IGNORECASE)
        print("Bearbeite Term :"+i)

        #TooMuchCounterInHere
        real_counter=0
        term_file_counter=0
        file_counter=0
        c_difference=0
        difference_counter=0
        counterInFile=0
        


        fileName_termFrequency_Dictionary = {}


        

        for file in glob.glob('Shakespeare/*.txt'):
            file_counter=file_counter+1
            inFile = False
            fileName=os.path.basename(file)
            #print(fileName)
            word_count=0
            with open(file) as fp:
                for result in pattern.findall(fp.read()):
                    counterInFile = counterInFile+1
                    real_counter = real_counter + 1
                    difference_counter=real_counter
            
                #Nochmal vom Anfang lesen    
                fp.seek(0)
                for line in fp:
                    word_count += len(line.split())
            
            if c_difference != difference_counter:
                term_file_counter= term_file_counter+1;
                c_difference = difference_counter
                #print(fileName+" : "+str(counterInFile)+ " Gesamtwörter: "+str(word_count))
                fileName_termFrequency_Dictionary[fileName]=counterInFile
                #termfrequenz=counterInFile/word_count
                termfrequenz=counterInFile
                #print("Die Termfrequenz ist:" + str(termfrequenz))
                #tf und filename in dictionary abspeichern :)
                if tf_boolean == True:
                    appending={(''+i+'','tf'):{(''+fileName+''): round(termfrequenz,6)}}
                    s1 = pd.DataFrame(main_dataframe)
                    s2 = pd.DataFrame(appending)
                    main_dataframe=s1.combine_first(s2)
                counterInFile=0
            else:
                fileName_termFrequency_Dictionary[fileName]=0
                if tf_boolean == True:
                    appending={(''+i+'','tf'):{(''+fileName+''): 0}}
                    s1 = pd.DataFrame(main_dataframe)
                    s2 = pd.DataFrame(appending)
                    main_dataframe=s1.combine_first(s2)
        
        
        
    
    
        
        #print(pattern1_string)
        #print("Der Begriff ist absolut "+str(real_counter)+ " mal vorhanden")
        #print("Ist in " +str(term_file_counter)+ " Dateien vorhanden")
        
        #print("df:" +str(term_file_counter))
        #print("Gesamter Dokumentenbestand: " +str(file_counter)+ " Dateien") #Dokumentenbestand am Anfang machen???
        
        #For every filename in FileOrdner append zu den einzelnen Möglichkeiten
        for file in glob.glob('Shakespeare/*.txt'):
                fileName=os.path.basename(file)
            
                #df appending
                if df_boolean == True:
                    appending= {(''+i+'','df'):{(''+fileName+''): term_file_counter}}
                    s1 = pd.DataFrame(main_dataframe)
                    s2 = pd.DataFrame(appending)
                    main_dataframe=s1.combine_first(s2)
            
                #idf appending
                try:
            
                    idf = math.log10(file_counter/term_file_counter)
                except ZeroDivisionError:
                    idf = 0
                appending= {(''+i+'','idf'):{(''+fileName+''): idf}}
                s1 = pd.DataFrame(main_dataframe)
                s2 = pd.DataFrame(appending)
                if idf_boolean == True:
                    main_dataframe=s1.combine_first(s2)
        
                #tf-idf appending
                if tf_idf_boolean == True:
                    try:
            
                        tf_idf = (1+math.log10(fileName_termFrequency_Dictionary[fileName]))*idf
                    except ValueError:
                        tf_idf = 0
                    appending= {(''+i+'','tf-idf'):{(''+fileName+''): tf_idf}}
                    s1 = pd.DataFrame(main_dataframe)
                    s2 = pd.DataFrame(appending)
                    main_dataframe=s1.combine_first(s2)
        
                #frequency appending
                if frequency_boolean == True:
                    appending= {(''+i+'','frequency'):{(''+fileName+''): fileName_termFrequency_Dictionary[fileName]}}
                    s1 = pd.DataFrame(main_dataframe)
                    s2 = pd.DataFrame(appending)
                    main_dataframe=s1.combine_first(s2)
        
                #Binär appending
                if binary_boolean == True:
                
                    local_binary=0
                    if fileName_termFrequency_Dictionary[fileName]>0:
                        local_binary=1
                    appending= {(''+i+'','binary'):{(''+fileName+''): local_binary }}
                    s1 = pd.DataFrame(main_dataframe)
                    s2 = pd.DataFrame(appending)
                    main_dataframe=s1.combine_first(s2)
        
    
    return main_dataframe


character_List= ["Antony","Brutus","CÆsar","Calphurnia","Cleopatra","mercy","worser"]
#character_List2= ["CÆsar"]
#list_of_searchTerms,tf_boolean,df_boolean,idf_boolean,tf_idf_boolean,frequency_boolean,binary_boolean)
main_dataframe=document_analyser(character_List,False,False,False,True,True,True)

#Ausgabeformate:

#pd.DataFrame(main_dataframe)
main_dataframe.to_excel("custom_output.xlsx")
#main_dataframe.to_csv("custom_output.csv")
print('Done!')
```
